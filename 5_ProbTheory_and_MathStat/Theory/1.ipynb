# Случайные события. Условная вероятность. Формула Байеса. Независимые испытания.

**Достоверное событие** - обязательно произойдёт в результате испытания

**Невозможное событие** - не может произойти в результате испытания

## Экспериментальные характеристики

**Относительная частота** - отношение числа появления события к общему числу проведенных испытаний

$ W(A) = \frac{m} {n} $

m - число появлений события А

n - общее число испытаний

 **Статистическая вероятность** - относительная частота при достаточно большом количестве испытаний

 $ P(A) = \frac {m}{n} $

 ## Классическая вероятность

 если заранее известны все вероятные исходы и они равновозможны, то можно воспользоваться классическим определением вероятности.

 **Вероятность события** - отношение числа элементарных исходов, благоприятствующих данному событию, к числу всех равновозможных исходов опыта

 $ P(A) = \frac {m}{n} $ 


## Комбинаторика

**Сочетания** - набор, состоящий из k элементов, выбранных из множества, состоящего из n различных элементов. Т.о. нас интересует не весь набор, а только его часть и нам не важен порядок.  

$ C_{n}^{k} = \frac {n!}{k!(n-k)!} $

**Перестановки** - комбинации из n элементов, отличающиеся порядком.

$ P_n = n!$

**Размещения** из k элементов, выбранных из множества n - комбинации, которые отличаются либо самими элементами, либо порядком их расположения

$ A_{n}^{k} = \frac {n!}{(n-k)!} $

## Зависимые и независимые события

**Зависимые события** - когда появление одного события влияет на появление другого события

$P(AB) = P(A) * P(B|A)$

**Независимые события** - когда появление одного из них не влияет на появление другого

$P(AB) = P(A) * P(B)$


## Совместные и несовместные события

**Несовместные события** - если поялвение одного из них исключает появление другого

**Совместные события** - если появление одного из них не исключает появление другого



## Произведение и сумма событий

**Произведение** - одновременное наступление событий A и B

**Сумма** - наступление события А или В (строго одно из них)

Для совместных событий:

$P(A*B) = P(A) * P(B)$

$P(A+B) = P(A) + P(B) - P(AB)$

Для несовместных событий:

$P(AB) = 0$

$P(A+B) = P(A) + P(B)$

## Формула полной вероятности

Если событие А может наступить только при появлении событий B1, B2, ... Bn образующих полную группу несовместных событий (т.е. при отдельном испытании обязательно возникнет только одно из них), то вероятность события А вычислятеся по формуле полной вероятности:

$P(A) = P(B1)*P(A|B1) + P(B2)*P(A|B2)+ ... + P(Bn) * P(A|Bn)$

## Формула Байеса

Для определения вероятности события В при условии, что событие А уже произошло

$P(B|A) = \frac {P(B) * P(A|B)}{P(A)}$

$P(B|A)$ - апостериорная, т.е. определенная после того, как произошло событие

$ P(В) $ - априорная вероятность - т.е. известная до проведения испытания

# Дискретные случайные величины. Закон распределения вероятностей. Биномиальный закон распределения. Распределение Пуассона

**случайная величина** - величина, принимающая в результате испытания одно возможное значение

дискретная случаная величина - принимает счетные значения

непрерывная случайная величина - лежит в некотором интервале, можем измерить с точностью, которую допускает измерительный прибор

**закон распределения вероятностей случайной величины** - соответствие между возможными значениями этой величины и вероятностями, с которыми эти значения могут появляться. Т.о. ЗРВ СВ - модель, описывающая поведение данной случайной величины.

## Биномиальное распределение

Случайная дискретная величина Х имеет биномиальное распределение с переменными k, n, p, если вероятность её распределения задается следующим уравнением (формула Бернулли):

$ P_n (X=k) = C_n^kp^kq^{n-k} $

где p - вероятность наступления события А в независимых испытаниях

q = 1 - p

k - число наступлений события


## Распределение Пуассона

Частный случай Биномиального распределения. 

Применяется в том случае, если проводится большое количество экспериментов, при этом вероятность наступления события А в каждом отдельном испытании мала (<0.1). 

Тогда для нахождения вероятности наступления события m раз за фиксированную единицу измерения применяют формулу Пуассона:

$ P_m \approx \frac {\lambda^m}{m!} * e^{-\lambda}$

$ \lambda $ - среднее количество наступлений события за определенную единицу измерения

Распределение Пуассона показывает вероятность числа наступлений события А за фиксированную единицу измерения

## Описательная статистика

### Математическое ожидание и дисперсия для биномиального распределения

**Математическое ожидание** - среднее значение случайной величины при числе испытаний стремящихся к бесконечности

$M(X) = np$

**Дисперсия** - характеризует степень рассеянности значений случайной величины относительно её математического ожидания

$ D(X) = npq $

**СКО** - среднеквадратическое отклонение показывает насколько далеко наблюдения могут быть разбросаны относительно их среднего значения

$ \sigma = \sqrt {D(X)} $

В пределах +/- одного интервала среднекдвадратического отклонения лежит 75% значений случйной величины

### Математическое ожидание и дисперсия для распределения Пуассона

$ M(X) = \lambda $

$ D(X) = \lambda $

### Генеральная совокупность и выборка

генеральная совокупность - множество, содержащее данные обо всех объектах, соответствующих определенным характеристикам. Сплошной контроль

выборка - случаным образом выбранная часть генеральной совокупности

по выборке мы получаем точечную оценку на основании которой делаем вывод обо всей генеральной совокупности

### Математическое ожидание 

среднее значение СВ при стремлении количества выборок или количества измерений к бесконечности

### Оценка математического ожидания 

среднее арифметическое одномерной случайной величины полученного в результате конечного числа испытаний

$ \overline M(X) = \frac {1}{n}\sum \limits_{i=1}^n x_i $

*математическое ожидание крайне чувствительно к выбросам*

Математическое ожидание используется для симметричных распределений. При несимметричных распределениях используются другие меры центральной тенденции

### среднее квадратичное (стандартное) отклонение

показывает насколько далеко наблюдения могут быть разбросаны относительно среднего значения. характеристика генеральной совокупности, на практике так же как и с математическим ожилданием измеряется по выборке - S - среднее квадратическое отклонение выборки.

### дисперсия

$ D = \sigma^2 = \frac {\sum\limits_{i=1}^{m}(x_i - \overline x)^2}{m}$

$ S^2 = \frac {\sum\limits_{i=1}^{n}(x_i - x)^2}{n-1}$

несмещенная оценка дисперсии по выборке - более точная оценка дисперсии. мы сознательно расширяем значение дисперсии (стандартного отклонения), потому что априори СКО и дисперсия генеральной совокупности выше чем СКО и дисперсия, полученные по значениям некоторой выборки

$ S^2 = \frac {\sum\limits_{i=1}^{n}(x_i - x)^2}{n}$

смещенная оценка дисперсии по выборке

n - объем выборки. 

### медиана 

мера центральной тенденции, использующаяся для несимметричных наблюдений

значение, которое делит выборку на две части, так что значения, которые меньше медианы, составляют половину выборки и значения, которые больше медианы также составляют половину выборки. 

### мода

наиболее часто встречающийся элемент выборки

### Квартили

первый квартиль - значение, такое что 25 % выборки не превышают данного значения
второй квартиль - синоним медианы - такое значение выборки, что 50 % её элементов не превышают данного значения
третий квартиль - значение, такое что 75 % выборки не превышают данного значения

интерквартильное расстояние - отрезок, равный разности 3 и 1 квартиля

### Персентили

если n * k / 100 - целое число, то k-й персентиль - среднее значение между номерами n * k / 100 и n * k / 101

если n * k / 100 - не целое число, то k-й персентиль - значение выборка[j + 1], где j - максимальное целое число < n * k / 100

### Размах

разница между максимальным и минимальным значениями

### Усеченное среднгее - статистическая мера центральной тенденции, рассчитанная как среднее значение для имеющегося набора данных, из которого исключены k% наибольших и k% наименьших значений. Как правило, процент удаляемых значений устанавливается в диапазоне 5-25%. 

# Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема.

### Непрерывная случайная величина

величина, принимающая все возможные значения, содержащиеся на промежутке, причем этот промежуток может быть как конечным, так и бесконечным.

### Функция распределения вероятностей

$F(x)$ - показывает для каждого значения x какова вероятность того, что случайная величина меньше либо равна х

### Плотность распределения вероятностей

$f(x) = F'(x)$

### Нормальное распределение (Распределение Гаусса)

Распределение вероятностей непрерывной случайной величины X, плотность распределения вероятностей которой описывается законом: 

$ f(x) = \frac {1}{\sigma \sqrt {2\pi}}e^-{\frac{(x-\mu)^2}{2\sigma^2}} $

$\mu = M(X)$

$\sigma^2 = D(x)$

### Правило трех сигм

- на отрезке от -$\sigma$ до $\sigma$ расположено около 68% наблюдений. также в точках +$\sigma$ и $-\sigma$ колокол имеет перегиб
- на отрезке от -$2\sigma$ до $2\sigma$ расположено около 95.4% наблюдений
- на отрезке от -$3\sigma$ до $3\sigma$ расположено около 99.72% наблюдений

### Стандартное нормальное распределение

$D(X)$ = 1
$\mu$ = 0

### Z - значение

показывает в скольких сигмах от математического ожидания лежит рассматриваемое значение случайной величины

$z = \frac{x - \mu}{\sigma}$

по значению z по таблицам определяется вероятность найти значения выше или ниже рассматриваемого значения

### Теорема
Если имеем СВ имеющую нормальное распределение с параметрами $\mu$ и $\sigma$, то СВ $z = \frac {X - \mu}{\sigma}$ будет иметь стандартное нормальное распределение

### Центальная предельная теорема

Пусть генеральная совокупность имеет любое распределение со средним арифметическим $\mu$ и дисперсией $\sigma^2$. Тогда закон распределения случайной величины представляющей собой средние выборочные, полученные из генеральной совокупности будет стремиться к нормальному с параметрами $\mu$ и $\frac{\sigma^2}{n}$, где n - мощность выборок

- совпадение среднего, моды и медианы

### Равномерное распределение непрерываной случайной величины
СВ равномерно распределена на отрезке [a, b], если плотность распределения вероятностей этой СВ равна 0 за пределами отрезка и $\frac {1}{b-a}$ внутри него

# Проверка статистических гипотез. P-значения. Доверительные интервалы. А/В тестирование

*Лекция 5. Проверка статистических гипотез. Вадим Леонардович Аббакумов. 2016. CSC*

- Проверку статистических гипотез на основе выборочных статистических данных называют статистической проверкой гипотез
- по результатам проверки статистических гипотез правильно говорить: **основная гипотеза отвергнута или основная гипотеза не отвергнута**. Так как обычно проверяется лишь достаточное условие. 
- ошибка первого рода состоит в том, что отвергается основная гипотеза при том, что на самом деле она была верна. **Ошибка первого рода тяжелее по своим последствиям**. При проверке статистических гипотез исходят именно из этой практики. Задаемся некоторым уровнем вероятности ошибки первого рода и стремимся не превысить его. Долю ошибок первого рода ограничивают сверху числом, называемым уровнем значимости. 
- ошибка второго рода состоит в том, что отвергается конкурирующая гипотеза при том, что она на самом деле верна. Уменьшить сложно. Нужно использовать "состоятельные" статистические критерии - те, для которых при фиксированном уровне значимости, ошибка второго рода с ростом наблюдений стремится к нулю. 

**Алгоритм проверки статистических гипотез**
- имеется n наблюдений
- заранее задан уровень значимости $\alpha$
- задан статистический критерий (т.е. функция от наблюдений)
- проверяем все условия, при которых критерий будет работать
- найдено p-значение (p-value) (иногда переводится как значимость)
- если $p < \alpha$ - гипотезу отвергаем, $p > \alpha$ - гипотезу не отвергаем

$p - value$ - вероятность того, что статистика критерия будет больше статистики критерия на имеющихся данных. при предположении, что основная гипотеза верна. 
Для проверки СВ на то, что она подчиняется закону нормального распределения рекомендуется использовать *критерий Шапиро-Уилка (или Колмогорова-Смирнова*, если наблюдений слишком много и применение первого критерия затруднительно)

А нужно ли проверять гипотезу нормальности? Исследования показывают, что если данные несущественно отличаются от нормальных, можно применять методы, применяемые к СВ с нормальным распределением. Что значит несущественно: 
- нет выбросов в данных
- нет явной асимметрии в гистограмме
- нет очень сильных отклонений формы гистограммы от колоколообразной формы

### Сравнение центров распределений 

**Центр распределения** - единственное число, которое описывало бы выборку. В качестве центра распределения чаще всего использют среднее арифметическое, медиану или усеченное среднее. 

Если центром распределения выбрана медиана, то центры двух распределений сравниваются методом *Манна-Уитни-Вилкоксона*. Если центром распределения определено среднее арифметическое, то центры сравниваются с помощью *критерия Стьюдента.* (варианты: Fligner-Killen, Brown-Forsythe) 

*Степик. Основы статистики*.
- Меры центральной тенденции: выборочное среднее, мода, медиана
- Меры изменчивости: размах,  дисперсия, стандартное отклонение

**Статистическая гипотеза** - предположение о неизвестном распределении СВ, соответствующих каким-либо представлениям о том явлении, которое изучается. 

Нулевая гипотеза - утверждение о свойствах генеральной совокупности, которое кажется правдоподобным, но требует проверки. Делает предположение о том, что различия между сравниваемыми выборками отсутствуют. Математический смысл заключается в том, что $\mu_1 = \mu_2$, т.е. различие между выборками стремиться к нулю. 

Альтернативная гипотеза - любая действительная гипотеза, отличная от нулевой. гипотеза, противостоящая нулевой. смысл заключается в том, что различия между выборками есть и они статистически достоверны. 

**Z-тест**

Применяется при известном СКО генеральной совокупности.

Z - критерий: $Z_н = \frac {\overline X - \mu}{\frac {\sigma}{\sqrt{n}}}$

$ \overline X $ - выборочное среднее

$ \mu $ - математическое ожидание генеральной совокупности

$ \sigma $ - СКО генеральной совокупности

$ n $ - объем выборки


*Мощность теста* - вероятность принять гипотезу $H_1$ при том, что она на самом деле верна. Устанавливается заранее. Мощность теста определяется обычно на уровне 80-90 % и участвует в определении необходимого объема выборки. 

*$\alpha$* - уровень значимости, вероятность ошибки первого рода, т.е. отклонение гипотезы $H_0$ при том, что она на самом деле верна. 

*$p-value$* - вероятность в распределении тест-статистики, которая лежит за пределами наблюдаемого значения (т.е. площадь под графиком плоности распределения вероятности ограниченная наблюдаемым значением)
Если $p-value > \alpha$ - нулевая гипотеза не может быть отвергнута
Если $p-value < \alpha$ - нулевая гипотеза отвергается

Нестрогое, но полезное определение: p-value - вероятнось того, что верна нулевая гипотеза.

Доверительный интервал показывает то, с какой вероятностью он захватит истинное значение оцениваемого параметра. 

Если известна дисперсия генеральной совокупности: 

$\overline X \pm z_{\alpha/2}*\sigma/\sqrt{n} $

Если неизвестная дисперсия генеральной совокупоности, то используем критерий стьюдента: 

$\overline X \pm t_{\alpha / 2} * \frac {se} {\sqrt{n}}$

se - несмещенная оценка дисперсии выборки


# Взаимосвязь величин. Параметрические и непараметрические показатели корреляции. Корреляционный анализ

**Корреляция** - математический показатель, по которому можно судить о наличии статистической взаимосвязи между двумя и более случайными величинами. 
**Коэффициент корреляции** - коэффициент, показывающий насколько велика линейная взаимосвязь между величинами

# Дисперсионный анализ. Логистическая регрессия 

используется для исследования влияния одного или нескольких качественных показателей на количественный показатель

### однофакторный дисперсионный анализ

на одну количественную переменную Y влияет один фактор (один качественный показатель), наблюдаемый на k уровнях (т.е. имеем к выборок для переменной Y)

ANOVA - analisys of variance - не покажет между какими именно уровнями качественного показателя найдены статистические различия. Ответ на этот вопрос могут дать post-hoc тесты

В дисперсионном анализе используется критерий Фишера (асимметричное распределение с удлиненным правым хвостом)

Для применения дисперсионного анализа дисперсии должны быть однородны (равны: Барлет-тест), подгруппы распределены нормально, независимость выборок

**Критерий Фишера**

$ F_H = \frac {\sigma_F^2}{\sigma_{ост}^2} $ - расчетное значение критерия Фишера

$ \sigma_F^2 $  - факторая дисперсия

$ \sigma_{ост}^2 $ - остаточная дисперсия

$ \sigma_F^2 = \frac {S_F^2}{k-1} $

$ \overline Y = \frac {1}{n} \sum \limits_{i=1}^{k}\overline y_i * n_i $ 

$ S_F^2 + S_{ост}^2 д.б. = S^2 = \sum \limits_{i=1}^{k} \sum \limits_{j=1}^{n_i}(y_{ij} - \overline Y)^2$ - сумма квадратов отклонений наблюдений от общего среднего

$ S_F $ - сумма квадратов отклонений средних групповых значений от общего среднего значения $ \overline Y $

$ S_F^2 = \sum\limits_{i=1}^k(\overline y_i - \overline Y) ^ 2n_i$

$ S_{ост}^2 =  \sum \limits_{i=1}^{k} \sum \limits_{j=1}^{n_i}(y_{ij} - \overline y_i)^2$ - остаточная сумма квадратов отклонений

$ \sigma_{ост}^2 = \frac {S_{ост}^2}{n-k} $

k - число подгрупп

k - 1 И n - k - степени свободы
